Built-in > Global > Local namespace
- Built-in namespace : 파이썬에서 제공되는 내장함수, Error/Exception의 이름 관리, 사용자는 Read만 가능
- Global namespace : 함수 외부에서 선언된 이름 관리
- Local namespae : 함수 내부에서 선언된 이름 관리
- Read: 명령의 위치에서의 namespace부터 검색, 있으면 사용 없으면 상위 namespace로 이동하여 검색, Built-in에도 없으면 NameError 발생
- Write: 명령의 위치에 해당하는 namespace에 등록 또는 수정

#### 상속(Inheritance)
- 부모 클래스의 속성을 자식 클래스가 사용할 수 있음

````python
class Student:
    def __init__(self, name, score):
        self.name = name          # 속성
        self._score = score       # 보호된 속성 (캡슐화) // 강제성 없음

    # 캡슐화: getter
    def get_score(self):
        return self._score

    # 캡슐화: setter
    def set_score(self, value):
        if value < 0 or value > 100:
            print('잘못된 점수입니다')
        else:
            self._score = value

    # 행동(메서드)
    def introduce(self):
        print(f'이름: {self.name}, 점수: {self._score}')

class GraduateStudent(Student):
    def __init__(self, name, score, thesis):
        super().__init__(name, score)
        self.thesis = thesis

    # 메서드 오버라이딩
    def introduce(self):
        print(f'대학원생: {self.name}, 점수: {self._score}, 논문: {self.thesis}')

s1 = Student('철수', 90)
s2 = GraduateStudent('영희', 95, 'AI 연구')

s1.introduce()
s2.introduce()
````

충돌 (Conflict): 같은 메서드가 있을 때
 > Python은 MRO (Method Resolution Order) 사용

````python
class A:
    def hello(self):
        print('A')

class B:
    def hello(self):
        print('B')

class C(A, B):
    pass

c = C()
c.hello()
print(C.__mro__)
````

- iterable은 iter(iterable)을 여러 번 실행할 수 있음
- 실행할 때마다 다른 객체 생성

````python
a = [1, 2, 3]
b = iter(a)
c = iter(a)
print(id(b), id(c))  # id가 다르므로 다른 객체이다.
# iterable은 출력을 위해 iterator를 사용해 문자열을 생성할 수 있다
# iterator는 출력을 위해 문자열을 생성할 수 없다.
# 문자열을 생성하려면 next를 사용한 객체 반환이 필요하기 때문
print(a, b, c)
````

- iterator는 next(iterator)를 통해 반복 실행하다가 StopIteration가 발생하면 더이상 사용하지 못함

````python
a = [1, 2, 3]
b = iter(a)
print(next(b))
print(next(b))
print(next(b))
print(next(b))
print(next(b))
````

#### argument의 종류
- positional : func(객체) - 순서에 따라 전달됨
- keyword : func(이름=객체) - 객체를 전달받을 parameter를 지정, 순서에 상관 없음
- p와 k를 함께 사용할 때는 항상 p가 앞, k가 뒤에 나와야 함

````python
# a, b, c는 non-default argument임
def func(a, b, c):
    print([a, b, c])

func(10, 20, 30)  # positional argument
func(a=30, b=10, c=20) # keyword argument
func(c=30, a=10, b=20) # keyword argument
func(10, c=30, b=20) # p와 k를 섞어 사용
````

#### parameter 종류
  - positional or keyword: parameter 앞에 *, 뒤에 ,/ 없음
  - positional only : parameter 뒤에 ,/ 가 있음
  - keyword only : parameter 앞에 *, 가 있음
  - var positional : parameter의 첫한글자가 *이다., 0개 이상의 argument를 묶어 1개의 tuple로 전달함
  - var keyword : parameter의 첫두글자가 **이다., 0개 이상의 argument를 묶어 1개의 dict로 전달함
  - var positional, var keyword 는 default argument 사용할 수 없음
  - var keyword 뒤에는 다른 parameter 사용할 수 없음

````python
# keyword only parameter
def func(*, a):
    print(a)

#func(10) # positional argument 사용 금지
func(a=10)
````

````python
# var positional parameter
def func(*a):
    print(a, type(a))

func()
func(1)
func(1, 2, 3)
````

````python
# var keyword parameter
def func(**a):
    print(a, type(a))

func()
func(A=1)
func(A=1, B=2)
#func(1='A') # parameter로 사용할 것이라서 이름규칙을 만족해야 함
````

#### lambda 함수 -> expression
```
lambda x : x**2
```

````python
# 숫자를 1개 입력 받아 제곱하는 함수
func = lambda x : x**2
func(2)
````

````python
# 숫자를 1개 입력 받아 '짝수' 또는 '홀수'를 반환하는 함수
func = lambda x : '짝수' if x % 2 == 0 else '홀수'
print(func(10))  # -> 짝수
print(func(5)) # -> 홀수
````

````python
# 숫자 2개를 입력 받아 합을 반환하는 함수
def func(x, y):
    return x + y

func = lambda x, y : x + y
func(10, 20)
````

````python
# a, b의 각 item을 더한 값을 item으로 하는 리스트 c 생성
# map, lambda 사용
a = [1, 2, 3, 4, 5]
b = [2, 4, 6, 8, 10]
#c = [3, 6, 9, 12, 15]
c = list(map(lambda x, y: x + y, a, b))
print(c)
````

````python
# a, b의 각 item을 더한 값을 item으로 하는 리스트 c 생성
# map 사용하지 않고 comprehension + zip을 사용해서 풀이
a = [1, 2, 3, 4, 5]
b = [2, 4, 6, 8, 10]
#c = [3, 6, 9, 12, 15]
c = [sum(x) for x in zip(a, b)]
c = [x[0] + x[1] for x in zip(a, b)]
c = [x + y for x, y in zip(a, b)]
c = [(lambda x, y: x + y)(x, y) for x, y in zip(a, b)]
print(c)
````

**enumerate(iterable, start=0)**
- iterator
- 정수와 item을 tuple형태로 하나씩 반환
- 정수는 기본 0부터 1씩 증가하는 숫자를 사용하지만 start를 사용해 시작 숫자를 변경할 수 있음

````python
#help(enumerate)
#  |  enumerate(iterable, start=0)

data = "ABCDE"
for x in data:
    print(x)

data = [1, 2, 3, 4, 5]
for i in range(len(data)):
    data[i] = data[i] ** 2
print(data)
````

````python
data = [1, 2, 3, 4, 5]
for i, d in enumerate(data, start=1):
    #data[i] = d ** 2
    print(f'{i}번째 데이터는 {d}입니다.')
#print(data)
````

formatting
- % 연산자 : 서식('%d' '%s' '%f' '%c')과 함께 서식 문자열 생성, expression을 % 연산의 오른쪽에 사용
- str.format() : 문자열의 format 메서드, placeholder({})와 함께 사용, expression을 format메서드의 argument로 전달
- f'' : f-string, placeholder와 함께 사용, expression을 placeholder 내부에 표기
- placeholder 내부에 서식(format)을 표기함(선택적 사용)

````python
a = 10
b = 20

# 10 + 20 = 30
print(a, '+', b, '=', a+b)
print('%d + %d = %d' % (a, b, a+b))
print('{} + {} = {}'.format(a, b, a+b))
print(f'{a} + {b} = {a+b}')
````

````python
num = 123.4567
print('%.1f' % num)
print('{:.1f}'.format(num))
print(f'{num:.1f}')
#
````

````python
data = ['바나나', 1000]
# 바나나는 1000원입니다.

a = '%s는 %d원입니다.' % (data[0], data[1])
b = '{}는 {}원입니다.'.format(*data)
c = f'{data[0]}는 {data[1]}원입니다.'
print(a, b, c, sep='\n')
````

````python
data = ['바나나', 1000]
# 1000원짜리 바나나!
a = '{1}원짜리 {0}!'.format(*data)
print(a)
````
